```python
import mujoco_py
import gym
from sb3_contrib import TQC #https://sb3-contrib.readthedocs.io/en/master/modules/tqc.html
```

#Create Environment

```python
#Play HumanoidStandup by random
env = gym.make('HumanoidStandup-v2')
env = gym.wrappers.RecordVideo(env, '/content/drive/MyDrive/Intro RL/Demo/TrialHunamnoid')
state = env.reset()
print(state.shape)
while True:
  action = env.action_space.sample()
  state, reward, done, info = env.step(action)
  if done:
    break;
env.close()
```

#Create a Deep Reinforcemant Learning Model

```python
model = TQC("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=3000000,)
```

```python
model.save("/content/drive/MyDrive/Intro RL/Demo/TrialHunamnoid/HumanoidStandup-Model02")
```
